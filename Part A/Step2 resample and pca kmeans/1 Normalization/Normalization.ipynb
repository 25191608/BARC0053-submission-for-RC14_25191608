```python
import pandas as pd
import numpy as np

def normalize_columns(df, exclude_cols):
    """
    Standardize all columns in the DataFrame except exclude_cols to the range [-1, 1].
    """
    # 1. Dynamically obtain the column names to be processed
    # Logic: The column name is not in the exclusion list, and the data type of the column is numeric (number).

    target_cols = [col for col in df.columns 
                   if col not in exclude_cols 
                   and pd.api.types.is_numeric_dtype(df[col])]
    
    print(f"Columns requiring standardization were detected.: {target_cols}")

    # 2. Iterate through each column and process it.
    for col in target_cols:
        col_data = df[col]
        
        # Calculate the maximum and minimum values
        _min = col_data.min()
        _max = col_data.max()
        
        # 3. To prevent division by zero (i.e., all values ​​in the column are the same).
        if _max == _min:
            print(f"Note: If all values ​​in column '{col}' are identical, skip normalization or set them to 0.")
            df[col] = 0 # Or maintain the original value as needed.
        else:
            # 4. Execute standardized formula: 2 * (x - min) / (max - min) - 1
            df[col] = ((col_data - _min) / (_max - _min)) * 2 - 1

    return df



# Read data
file_path = "input/Combine 1.csv"  # 替换为你的文件路径
df = pd.read_csv(file_path)

# Define columns that do not need to be processed（id, row_index, col_index）
exclusions = ['id', 'row_index', 'col_index']

# Calling functions
df_normalized = normalize_columns(df, exclusions)


print(df_normalized.head())


df_normalized.to_csv('Normalized_Data_only_Laplacian.csv', index=False)
```
